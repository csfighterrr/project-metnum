```python
# Install dependencies as needed:
# pip install pandas

# Import library yang dibutuhkan
import pandas as pd # Hanya untuk membaca CSV dari file

# --- Bagian 0: Membaca Data CSV dari Google Drive ---

# Mounting Google Drive
# Ini adalah bagian yang harus kamu jalankan di Google Colab
# Akan muncul prompt untuk otorisasi akses Google Drive
try:
    from google.colab import drive
    drive.mount('/content/drive')
    print("Google Drive berhasil di-mount.")
except ImportError:
    print("Peringatan: Tidak di lingkungan Google Colab. Pastikan file CSV tersedia di direktori lokal.")
    # Jika tidak di Colab, asumsikan file ada di direktori yang sama dengan script
    # Anda bisa menyesuaikan path di bawah ini jika file ada di lokasi lain.
    pass # Lanjutkan saja, asumsikan file_path_in_drive akan mengarah ke lokal jika tidak di Colab

# Tentukan path file CSV di Google Drive Anda
# Sesuaikan path ini dengan lokasi file Laptop_price.csv di Google Drive Anda
# Contoh: '/content/drive/MyDrive/FolderDatasetSaya/Laptop_price.csv'
file_path_in_drive = '/content/drive/MyDrive/Dataset/Laptop_price.csv' # SESUAIKAN PATH INI!

print(f"Mencoba membaca dataset dari: {file_path_in_drive}")
try:
    df = pd.read_csv(file_path_in_drive)
    print("Dataset berhasil dimuat.")
    print("Beberapa record pertama:\n", df.head())
except FileNotFoundError:
    print(f"Error: File '{file_path_in_drive}' tidak ditemukan.")
    print("Pastikan Anda sudah me-mount Google Drive dan path file sudah benar.")
    exit()
except Exception as e:
    print(f"Terjadi kesalahan saat memuat dataset: {e}")
    exit()

# Definisi fitur (X) dan target (y)
# Kita akan menggunakan semua fitur numerik kecuali 'Brand' sebagai fitur (X)
# dan 'Price' sebagai target (y).
feature_columns = ['Processor_Speed', 'RAM_Size', 'Storage_Capacity', 'Screen_Size', 'Weight']
target_column = 'Price'

# Ekstraksi data fitur dan target ke dalam list Python biasa (Manual)
# Ini penting untuk menunjukkan bahwa setelah pembacaan, semua operasi numerik adalah manual.
data_X_original = []
data_y_original = []

for index, row in df.iterrows():
    features_row = []
    for col in feature_columns:
        features_row.append(row[col])
    data_X_original.append(features_row)
    data_y_original.append(row[target_column])

# Jumlah sampel data dan jumlah fitur
n_samples = len(data_X_original)
n_features = len(feature_columns)

print(f"\nJumlah sampel data: {n_samples}")
print(f"Jumlah fitur: {n_features}")
print("Data fitur (X) yang diekstrak (5 baris pertama):\n", [x for x in data_X_original[:5]])
print("Data target (y) yang diekstrak (5 baris pertama):\n", [f"{y:.2f}" for y in data_y_original[:5]])

# --- Bagian 1: Fungsi Utilitas Dasar (Manual) ---

def calculate_mean(data):
    """Menghitung rata-rata dari sebuah list secara manual."""
    if not data:
        return 0.0
    total = 0.0
    for item in data:
        total += item
    return total / len(data)

def calculate_min(data):
    """Menghitung nilai minimum dari sebuah list secara manual."""
    if not data:
        return None
    min_val = data[0]
    for val in data:
        if val < min_val:
            min_val = val
    return min_val

def calculate_max(data):
    """Menghitung nilai maksimum dari sebuah list secara manual."""
    if not data:
        return None
    max_val = data[0]
    for val in data:
        if val > max_val:
            max_val = val
    return max_val

# --- Bagian 2: Penskalaan Data (Min-Max Scaling) untuk Multiple Features (Manual) ---

# Menyimpan nilai min dan max untuk setiap fitur dan target
# Akan ada satu list untuk min_vals_X dan max_vals_X, dan satu nilai untuk min_y, max_y
min_vals_X = [0.0] * n_features
max_vals_X = [0.0] * n_features
min_y = data_y_original[0]
max_y = data_y_original[0]

# Inisialisasi min/max untuk fitur
for j in range(n_features):
    min_vals_X[j] = data_X_original[0][j]
    max_vals_X[j] = data_X_original[0][j]
    for i in range(n_samples):
        if data_X_original[i][j] < min_vals_X[j]:
            min_vals_X[j] = data_X_original[i][j]
        if data_X_original[i][j] > max_vals_X[j]:
            max_vals_X[j] = data_X_original[i][j]

# Inisialisasi min/max untuk target y
for i in range(n_samples):
    if data_y_original[i] < min_y:
        min_y = data_y_original[i]
    if data_y_original[i] > max_y:
        max_y = data_y_original[i]

# Fungsi penskalaan dan deskalasi untuk nilai tunggal
def scale_value(val, min_val, max_val):
    if (max_val - min_val) == 0:
        return 0.0
    return (val - min_val) / (max_val - min_val)

def descale_value(scaled_val, min_val, max_val):
    return scaled_val * (max_val - min_val) + min_val

# Skala seluruh data X dan y
data_X_scaled = [[0.0 for _ in range(n_features)] for _ in range(n_samples)]
data_y_scaled = [0.0] * n_samples

for i in range(n_samples):
    for j in range(n_features):
        data_X_scaled[i][j] = scale_value(data_X_original[i][j], min_vals_X[j], max_vals_X[j])
    data_y_scaled[i] = scale_value(data_y_original[i], min_y, max_y)

print("\n--- Data yang Diskala (Beberapa Awal) ---")
print("Fitur X (scaled, 5 baris pertama):")
for i in range(min(5, n_samples)):
    print([f"{x:.4f}" for x in data_X_scaled[i]])
print("Target y (scaled, 5 baris pertama):", [f"{x:.4f}" for x in data_y_scaled[:5]], "...")
print("-" * 30)

# --- Bagian 3: Implementasi Regresi Linear Berganda dengan Gradient Descent (Manual) ---

# Inisialisasi koefisien (bobot) untuk setiap fitur dan bias
# Ada n_features bobot + 1 bias
coefficients = [0.0] * (n_features + 1) # coefficients[0] adalah bias (intercept)

learning_rate = 0.01 # Tingkat pembelajaran, perlu disesuaikan!
epochs = 5000        # Jumlah iterasi

print("\n--- Proses Pelatihan Gradient Descent ---")
for epoch in range(epochs):
    # Hitung prediksi dan error untuk setiap sampel
    predictions_scaled = []
    errors_scaled = [] # (y_actual - y_predicted)
    for i in range(n_samples):
        # Hitung prediksi (y_hat = bias + w1*x1 + w2*x2 + ...)
        y_predicted_scaled = coefficients[0] # Mulai dengan bias
        for j in range(n_features):
            y_predicted_scaled += coefficients[j+1] * data_X_scaled[i][j]
        predictions_scaled.append(y_predicted_scaled)
        errors_scaled.append(data_y_scaled[i] - y_predicted_scaled)

    # Hitung gradien untuk setiap koefisien
    # Gradien untuk bias
    grad_bias = 0.0
    for err in errors_scaled:
        grad_bias += err
    grad_bias = (-2.0 / n_samples) * grad_bias

    # Gradien untuk setiap bobot fitur
    grads_features = [0.0] * n_features
    for j in range(n_features):
        sum_grad_feature_j = 0.0
        for i in range(n_samples):
            sum_grad_feature_j += data_X_scaled[i][j] * errors_scaled[i]
        grads_features[j] = (-2.0 / n_samples) * sum_grad_feature_j

    # Perbarui koefisien
    coefficients[0] = coefficients[0] - learning_rate * grad_bias
    for j in range(n_features):
        coefficients[j+1] = coefficients[j+1] - learning_rate * grads_features[j]

    # Opsional: Cetak MSE setiap beberapa epoch untuk memantau konvergensi
    if epoch % (epochs // 10) == 0: # Cetak 10 kali selama pelatihan
        mse_scaled = 0.0
        for err in errors_scaled:
            mse_scaled += err**2
        mse_scaled /= n_samples
        print(f"Epoch {epoch}, MSE (scaled): {mse_scaled:.6f}")

print("\n--- Koefisien Akhir (Hasil Gradient Descent pada Data Diskala) ---")
print(f"Bias (Intercept): {coefficients[0]:.6f}")
for j, feature_name in enumerate(feature_columns):
    print(f"Bobot {feature_name}: {coefficients[j+1]:.6f}")
print("-" * 30)


# --- Bagian 4: Fungsi Prediksi dan Deskalasi (Manual) ---

def predict_price_multiple(features_raw, coefficients_model, min_vals_X, max_vals_X, min_y, max_y):
    """
    Prediksi harga menggunakan koefisien yang ditemukan,
    termasuk penskalaan input dan deskalasi output.
    """
    # 1. Skala fitur input
    features_scaled = []
    for j in range(len(features_raw)):
        features_scaled.append(scale_value(features_raw[j], min_vals_X[j], max_vals_X[j]))

    # 2. Hitung prediksi pada skala diskala
    predicted_scaled = coefficients_model[0] # Bias
    for j in range(len(features_scaled)):
        predicted_scaled += coefficients_model[j+1] * features_scaled[j]

    # 3. Deskalasi prediksi kembali ke skala original
    predicted_original = descale_value(predicted_scaled, min_y, max_y)
    return predicted_original

# --- Bagian 5: Evaluasi Model (R-squared) (Manual) ---

# Hitung rata-rata harga original untuk R-squared
mean_price_original = calculate_mean(data_y_original)

# Hitung SSR dan SST
sum_of_squares_residuals = 0.0 # SSR
total_sum_of_squares = 0.0     # SST

print("\n--- Evaluasi Model (R-squared) ---")
print("Harga Aktual | Harga Prediksi | Selisih Absolut")
print("-----------------------------------------------")
for i in range(n_samples):
    # Prediksi harga untuk sampel ke-i
    current_features = data_X_original[i]
    predicted_price_original = predict_price_multiple(
        current_features,
        coefficients,
        min_vals_X,
        max_vals_X,
        min_y,
        max_y
    )

    # Hitung komponen SSR
    sum_of_squares_residuals += (data_y_original[i] - predicted_price_original)**2

    # Hitung komponen SST
    total_sum_of_squares += (data_y_original[i] - mean_price_original)**2

    # Hanya tampilkan beberapa baris pertama/terakhir untuk keterbacaan
    if i < 5 or i >= n_samples - 5:
        print(f"{data_y_original[i]:<12.2f} | {predicted_price_original:<14.2f} | {abs(data_y_original[i] - predicted_price_original):.2f}")
    elif i == 5:
        print("...")

# Hitung R-squared
if total_sum_of_squares == 0:
    r_squared = 0.0 # Hindari pembagian nol jika semua nilai y sama
else:
    r_squared = 1.0 - (sum_of_squares_residuals / total_sum_of_squares)

print("-" * 60)
print(f"R-squared (Koefisien Determinasi): {r_squared:.4f}")
print("Semakin mendekati 1, semakin baik model memprediksi variabilitas harga.")
print("-" * 60)

# --- Bagian 6: Interaksi Pengguna untuk Prediksi Baru (Manual) ---

print("\n--- Coba Prediksi Harga Laptop Baru ---")
print("Masukkan nilai-nilai berikut untuk laptop baru:")
while True:
    try:
        user_input_features = []
        for j, feature_name in enumerate(feature_columns):
            input_val_str = input(f"   Masukkan {feature_name} (range: {min_vals_X[j]:.2f}-{max_vals_X[j]:.2f}): ")
            if input_val_str.lower() == 'exit':
                break # Keluar dari loop input fitur
            user_input_features.append(float(input_val_str))

        if len(user_input_features) != n_features: # Jika pengguna mengetik 'exit' di tengah input fitur
            if input_val_str.lower() == 'exit':
                break # Keluar dari loop utama
            else: # Ini seharusnya tidak terjadi jika semua input valid
                print("Jumlah input fitur tidak sesuai. Silakan coba lagi.")
                continue

        # Lakukan prediksi
        predicted_price_new = predict_price_multiple(
            user_input_features,
            coefficients,
            min_vals_X,
            max_vals_X,
            min_y,
            max_y
        )

        print(f"\nPrediksi harga untuk laptop dengan fitur tersebut adalah sekitar {predicted_price_new:.2f} Rupiah.")
        print("-" * 40)

    except ValueError:
        print("Input tidak valid. Harap masukkan angka atau 'exit'.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

print("\nProgram selesai. Terima kasih!")
```
